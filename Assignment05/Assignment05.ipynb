# College Scorecard 2017-2018 Data




# Read the file and print row and column information

```
import pandas as pd

college_data = pd.read_csv("MERGED2017_18_PP.csv")                  #Read the csv file, assign to college_data var

print("Number of colleges/rows: ",len(college_data))                #Print the number of colleges and variables
print("Number of variables/columns: ", len(college_data. columns))




```

# Create a smaller data frame with the variables:


1.   UNITID:            Unit ID for institution
2.   INSTNM:            Institution name
3.   CITY:              Institution city
4.   STABBR:            Institution state (abbreviation)
5.   ZIP:               Zip code
6.   ADM_RATE:          Institution overall admission rate   
7.   UGDS:              Enrollment of undergraduate certificate/degree-seeking students (size)
8.   TUITIONFEE_IN:     In-state tuition and fees
9.   MN_EARN_WNE_P10:   Mean earnings of students working and not enrolled 10 years after entry



```
subset_college_data = college_data[['UNITID','INSTNM','CITY','STABBR','ZIP','ADM_RATE','UGDS','TUITIONFEE_IN','MN_EARN_WNE_P10']]
subset_college_data
```

# Print the first, last, and a random sample of 10 colleges from the smaller data frame

```
subset_college_data.iloc[0:10]                                   #First 10 
```

```
subset_college_data.iloc[7102:]                                      #Last 10
```

```
subset_college_data.sample(n=10)                                     #Random 5
```

# Display UMBC, University of MD College Park, Loyola College of MD, and Johns Hopkins University

```
four_colleges_data=subset_college_data.loc[subset_college_data['INSTNM'] == 'University of Maryland-Baltimore County'].append(subset_college_data.loc[subset_college_data["INSTNM"]=='University of Maryland-College Park']).append(subset_college_data.loc[subset_college_data['INSTNM'] == 'Loyola University Maryland']).append(subset_college_data.loc[subset_college_data['INSTNM'] == 'Johns Hopkins University'])
four_colleges_data
```

# Calculate the min, max, and average in-state tuition for the four institutions above

```
four_colleges_data['TUITIONFEE_IN'].min()             #min
```

```
four_colleges_data['TUITIONFEE_IN'].max()             #max
```

```
four_colleges_data["TUITIONFEE_IN"].mean()            #average
```

# Display MD, VA, DC, DE, and PA institution data

```
midatl_colleges_data=subset_college_data.loc[subset_college_data['STABBR'] == 'MD'].append(subset_college_data.loc[subset_college_data["STABBR"]=='VA']).append(subset_college_data.loc[subset_college_data['STABBR'] == 'DC']).append(subset_college_data.loc[subset_college_data['STABBR'] == 'DE']).append(subset_college_data.loc[subset_college_data['STABBR'] == 'PA'])
midatl_colleges_data
```

# Save MD, VA, DC, DE, and PA data to a file

```
midatl_colleges_data.to_csv (r'Five_States_Colleges.csv', index = False, header=True)
print(midatl_colleges_data)
```
